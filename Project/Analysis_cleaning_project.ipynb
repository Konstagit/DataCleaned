```python
melb_df['Date']=pd.to_datetime(melb_df['Date'], format='%Y-%m-%d')
```

```python
melb_df['Date'].dt.quarter.count()
```

```python
unique_list=[]
for col in melb_df.columns:
    item=(col,melb_df[col].nunique(),melb_df[col].dtypes)
    unique_list.append(item)
unique_df=pd.DataFrame(unique_list,
                       columns = ['Column_Name', 'Num_Unique', 'Type']).sort_values(by='Num_Unique',ignore_index=True)
```

```python
display(unique_df)
```

```python
list_to_exclude=['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_val_counts=150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_val_counts and col not in list_to_exclude:
        melb_df[col]=melb_df[col].astype('category')
display(melb_df.info())
```
